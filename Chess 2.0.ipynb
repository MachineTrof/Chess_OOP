```python
#Класс-всеяотец "Фигурошный"
class figure(object):
    def __init__(self,colour):
        self.colour = colour #Хранит лишь один артибут - цвет. Только он будет наследоваться остальными классами
```

```python
#Класс-сын Пешка
class Peshka(figure):
    def __init__(self,colour):
        super().__init__(colour) #Наследуем цвет
        
    def pr(self): #Функция перевода из класса в строчное представление
        if self.colour == 'W':
            return 'P'
        else:
            return 'p'
        
    def Xod(self,figure,X,Y,Pole):
        global White_Turn_P, Black_Turn_P #Глобальные переменные, отвечающие за то, ходил ли прошлым ходом игрок на две клетки пешкой вперёд
        global LineW_Turn_P, LineB_Turn_P #Глобальные переменные, отвечающие за то, на какой линии подвигалась пешка на две клетки (под линиями я подрузамеваю A,B,...,H, конвертированные в цифры (1,2,...,8))
        konvert = {1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'F', 7:'G', 8:'H'} #Конвертация цифр в буквы
        Konvert = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8} #Конвертация букв в цифры
        #Такое разветвление (в отличие от других фигур) по двум причинам: Уникальные движения пешками (одни "вверх", другие - "вниз"), Уникальный набор замен пешек
        if figure == 'P':
            #Определение вражеских фигур
            VragAlph = 'prmsq'
            VozmXods = {}
            k = 1
            #Выявляем возможные ходы
            try:
                if ((Pole[X-1])[Y]).pr() == '-':
                    VozmXods[k] = [X-1,konvert[Y]]
                    k += 1
            except: 
                pass
            try:
                if ((Pole[X-1])[Y+1]).pr() in VragAlph:
                    VozmXods[k] = [X-1,konvert[Y+1]]
                    k += 1
            except:
                pass
            try:
                if ((Pole[X-1])[Y-1]).pr() in VragAlph:
                    VozmXods[k] = [X-1,konvert[Y-1]]
                    k += 1
            except:
                pass
            try:
                if X == 7 and (((Pole[X-1])[Y]).pr() == '-') and (((Pole[X-2])[Y]).pr() == '-'):
                    VozmXods[k] = [X-2,konvert[Y]]
                    White_Turn_P = True
                    k += 1
            except:
                pass
            try: #Проверка взятия на проходе
                if Black_Turn_P == True and X == 4:
                    if Y-1 == LineB_Turn_P:
                        VozmXods[k] = [X-1,konvert[Y-1]]
                        k += 1
                    elif Y+1 == LineB_Turn_P:
                        VozmXods[k] = [X-1,konvert[Y+1]]
                        k += 1
                    else:
                        pass
            except:
                pass
            if k == 1:
                print('Данной фигурой невозможно походить, выберите другую!')
                return
            #И даём выбор игроку
            print(VozmXods)
            Korrekt = False
            while Korrekt == False:
                try:
                    Vibor = int(input('Выберите, куда вы походите, введя нужную цифру = '))
                    if Vibor >= 1 and Vibor <= (k-1):
                        Korrekt = True
                        x = (VozmXods[Vibor])[0]
                        y = Konvert[(VozmXods[Vibor])[1]]
                    else:
                        print('Некорректный ввод выбора! Попробуйте ещё раз!')
                except:
                    print('Некорректный ввод выбора! Попробуйте ещё раз!')
            (Pole[X])[Y], (Pole[x])[y] = Nothing(), Peshka('W') #"Меняем местами" Пешку и Пустоту
            if White_Turn_P == True and x == 5:
                White_Turn_P = True
                LineW_Turn_P = y
            else:
                White_Turn_P = False
                LineW_Turn_P = -100
            if ((Pole[x+1])[y]).pr() == 'p' and y == LineB_Turn_P and Black_Turn_P == True:
                (Pole[x+1])[y] = Nothing() #Если произошло взятие на проходе, то убиваем пешку позади
            #Замена пешки, если она доковыляла до другого края
            if x == 1:
                Korrekt = False
                while Korrekt == False:
                    try:
                        Nf = str(input('Выберите замену пешки (R,M,S,Q) = '))
                        if Nf in 'RMSQ':
                            Korrekt = True
                        else:
                            print('Некорректный ввод замены! Попробуйте ещё раз!')
                    except:
                        print('Некорректный ввод замены! Попробуйте ещё раз!')
                if Nf == 'R':
                    (Pole[x])[y] = Rook('W')
                elif Nf == 'M':
                    (Pole[x])[y] = Mare('W')
                elif Nf == 'S':
                    (Pole[x])[y] = Slon('W')
                else:
                    (Pole[x])[y] = Queen('W')
        #"То же самое", что и сверху
        else:
            VragAlph = 'PRMSQ'
            VozmXods = {}
            k = 1
            try:
                if ((Pole[X+1])[Y]).pr() == '-':
                    VozmXods[k] = [X+1,konvert[Y]]
                    k += 1
            except:
                pass
            try:
                if ((Pole[X+1])[Y+1]).pr() in VragAlph:
                    VozmXods[k] = [X+1,konvert[Y+1]]
                    k += 1
            except:
                pass
            try:
                if ((Pole[X+1])[Y-1]).pr() in VragAlph:
                    VozmXods[k] = [X+1,konvert[Y-1]]
                    k += 1
            except:
                pass
            try:
                if X == 2 and (((Pole[X+1])[Y]).pr() == '-') and (((Pole[X+2])[Y]).pr() == '-'):
                    VozmXods[k] = [X+2,konvert[Y]]
                    Black_Turn_P = True
                    k += 1
            except:
                pass
            try:
                if White_Turn_P == True and X == 5:
                    if Y-1 == LineW_Turn_P:
                        VozmXods[k] = [X+1,konvert[Y-1]]
                        k += 1
                    elif Y+1 == LineW_Turn_P:
                        VozmXods[k] = [X+1,konvert[Y+1]]
                        k += 1
                    else:
                        pass
            except:
                pass
            if k == 1:
                print('Данной фигурой невозможно походить, выберите другую!')
                return
            print(VozmXods)
            Korrekt = False
            while Korrekt == False:
                try:
                    Vibor = int(input('Выберите, куда вы походите, введя нужную цифру = '))
                    if Vibor >= 1 and Vibor <= (k-1):
                        Korrekt = True
                        x = (VozmXods[Vibor])[0]
                        y = Konvert[(VozmXods[Vibor])[1]]
                    else:
                        print('Некорректный ввод выбора! Попробуйте ещё раз!')
                except:
                    print('Некорректный ввод выбора! Попробуйте ещё раз!')
            (Pole[X])[Y], (Pole[x])[y] = Nothing(), Peshka('b')
            if Black_Turn_P == True and x == 4:
                Black_Turn_P = True
                LineB_Turn_P = y
            else:
                Black_Turn_P = False
                LineB_Turn_P = -100
            if ((Pole[x-1])[y]).pr() == 'P' and y == LineW_Turn_P and White_Turn_P == True:
                (Pole[x-1])[y] = Nothing()
            if x == 8:
                Korrekt = False
                while Korrekt == False:
                    try:
                        Nf = str(input('Выберите замену пешки (r,m,s,q) = '))
                        if Nf in 'rmsq':
                            Korrekt = True
                        else:
                            print('Некорректный ввод выбора! Попробуйте ещё раз!')
                    except:
                        print('Некорректный ввод выбора! Попробуйте ещё раз!')
                if Nf == 'r':
                    (Pole[x])[y] = Rook('b')
                elif Nf == 'm':
                    (Pole[x])[y] = Mare('b')
                elif Nf == 's':
                    (Pole[x])[y] = Slon('b')
                else:
                    (Pole[x])[y] = Queen('b')
```
