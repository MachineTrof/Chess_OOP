Новые фигуры:
1.) Инквизитор. В начале игры стоит на месте ферзя, ходит, как он, однако способен уничтожать свои же фигуры, что может перевернуть ход игры как в свою, так и в чужую пользу. Открывается новое пространство для обдумывания и анализа
2.) Одержимый (или Фанатик). Самая слабая фигура из добавленных. Заменяет собой пешку перед ферзём. Почти целиком её и копирует, однако умеет бить фигуры перед собой и неуязвим к взятию на проходе, впрочем, и сам не умеет вылавливать прыткие пешки
3.) Могучие (или Великие). Особо умелые кавалеристы-рыцари, умеющие без труда заходить в тыл к врагу, к сожалению, всё также бесполезные в ближнем бою. Умеют скакать не только буквой "Г", но и на две клетки в стороны

```python
#Класс-всеяотец "Фигурошный"
class figure(object):
    def __init__(self,colour):
        self.colour = colour
```

```python
#Класс-сын Слон
class Slon(figure):
    def __init__(self,colour):
        super().__init__(colour)
        
    def pr(self):
        if self.colour == 'W':
            return 'S'
        else:
            return 's'
        
    def Xod(self,figure,X,Y,Pole):
        global White_Turn_P
        global Black_Turn_P
        konvert = {1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'F', 7:'G', 8:'H'} #Конвертация цифр в буквы
        Konvert = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8} #Конвертация букв в цифры
        if figure == 'S':
            VragAlph = 'oprmsi'
            coloure = 'W'
        else:
            VragAlph = 'OPRMSI'
            coloure = 'b'
        VozmXods = {}
        k = 1
        kontrol = True #Засекает, если кто-то мешает слону пройти от края до края поля
        Xt, Yt = X, Y
        while kontrol == True: #По диагонали вправо вверх
            Xt += 1
            Yt += 1
            try:
                if ((Pole[Xt])[Yt]).pr() == '-':
                    VozmXods[k] = [Xt,konvert[Yt]]
                    k += 1
                else:
                    if ((Pole[Xt])[Yt]).pr() in VragAlph:
                        VozmXods[k] = [Xt,konvert[Yt]]
                        k += 1
                    kontrol = False
            except:
                kontrol = False
        kontrol = True
        Xt, Yt = X, Y
        while kontrol == True: #По диагонали влево вниз
            Xt -= 1
            Yt -= 1
            try:
                if ((Pole[Xt])[Yt]).pr() == '-':
                    VozmXods[k] = [Xt,konvert[Yt]]
                    k += 1
                else:
                    if ((Pole[Xt])[Yt]).pr() in VragAlph:
                        VozmXods[k] = [Xt,konvert[Yt]]
                        k += 1
                    kontrol = False
            except:
                kontrol = False
        kontrol = True
        Xt, Yt = X, Y
        while kontrol == True: #По диагонали влево вверх
            Xt += 1
            Yt -= 1
            try:
                if ((Pole[Xt])[Yt]).pr() == '-':
                    VozmXods[k] = [Xt,konvert[Yt]]
                    k += 1
                else:
                    if ((Pole[Xt])[Yt]).pr() in VragAlph:
                        VozmXods[k] = [Xt,konvert[Yt]]
                        k += 1
                    kontrol = False
            except:
                kontrol = False
        kontrol = True
        Xt, Yt = X, Y
        while kontrol == True: #По диагонали вправо вниз
            Xt -= 1
            Yt += 1
            try:
                if ((Pole[Xt])[Yt]).pr() == '-':
                    VozmXods[k] = [Xt,konvert[Yt]]
                    k += 1
                else:
                    if ((Pole[Xt])[Yt]).pr() in VragAlph:
                        VozmXods[k] = [Xt,konvert[Yt]]
                        k += 1
                    kontrol = False
            except:
                kontrol = False
        if k == 1:
            print('Данной фигурой невозможно походить, выберите другую!')
            return
        print(VozmXods)
        Korrekt = False
        while Korrekt == False:
            try:
                Vibor = int(input('Выберите, куда вы походите, введя нужную цифру = '))
                if Vibor >= 1 and Vibor <= (k-1):
                    Korrekt = True
                    x = (VozmXods[Vibor])[0]
                    y = Konvert[(VozmXods[Vibor])[1]]
                else:
                    print('Некорректный ввод выбора! Попробуйте ещё раз!')
            except:
                print('Некорректный ввод выбора! Попробуйте ещё раз!')
        (Pole[X])[Y], (Pole[x])[y] = Nothing(), Slon(coloure)
        if Slon(coloure) == 'S': #Если предыдущим ходом ходила пешка, то мы заменим её ход на False
            White_Turn_P = False
        else:
            Black_Turn_P = False
```
